````python
# Chart - 2 visualization code
# Weekly Sales by month -> to see how sales vary month-to-month

df['Date'] = pd.to_datetime(df['Date'])

monthly_sales = df.groupby(df['Date'].dt.to_period('M'))['Weekly_Sales'].sum()

plt.figure(figsize=(10, 6))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title('Monthly Aggregate of Weekly Sales')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')
plt.show()
````

#### Chart - 4

````python
# Chart - 4 visualization code
# Holiday vs. Non-Holiday Weekly Sales Trends -> to see how trends depends on holidays

df['Date'] = pd.to_datetime(df['Date'])

df['Week'] = df['Date'].dt.isocalendar().week
avg_weekly_sales = df.groupby('Week')['Weekly_Sales'].mean()

avg_weekly_sales_by_holiday = df.groupby(['Week', 'IsHoliday'])['Weekly_Sales'].mean().unstack()
avg_weekly_sales_by_holiday = avg_weekly_sales_by_holiday.fillna(0)

plt.figure(figsize=(10, 6))
avg_weekly_sales.plot(label='Total Sales', color='black', linestyle='--')
avg_weekly_sales_by_holiday[True].plot(label='Holiday Sales', color='red')
avg_weekly_sales_by_holiday[False].plot(label='Non-Holiday Sales', color='blue')

plt.title('Weekly Sales Comparison: Holiday vs. Non-Holiday')
plt.xlabel('Week')
plt.ylabel('Average Weekly Sales')
plt.legend()
plt.grid(True)
plt.show()
````

##### 1. Why did you pick the specific chart?


   - This chart was chosen to compare trends in weekly sales between holiday and non-holiday weeks, providing insights into how holidays impact sales dynamics.







##### 2. What is/are the insight(s) found from the chart?

- The chart illustrates the average weekly sales trends, highlighting differences between total, holiday, and non-holiday sales.
- Observing peaks or dips in holiday weeks compared to non-holiday weeks provides insights into the impact of holidays on sales patterns.

##### 3. Will the gained insights help creating a positive business impact? 
Are there any insights that lead to negative growth? Justify with specific reason.


 - Yes, the insights from comparing holiday and non-holiday weekly sales can positively impact business decisions. Understanding the variations in sales during holiday periods allows for targeted marketing strategies, optimized inventory management, and improved resource allocation to capitalize on increased consumer activity during holidays.

#### Chart - 5

````python
# Chart - 5 visualization code
df['Date'] = pd.to_datetime(df['Date'])

df['Month'] = df['Date'].dt.month
df['Day_of_Week'] = df['Date'].dt.day_name()

pivot_table = df.pivot_table(index='Month', columns='Day_of_Week', values='Weekly_Sales', aggfunc='mean')

plt.figure(figsize=(10, 6))
plt.imshow(pivot_table, cmap='viridis', interpolation='nearest')

plt.title('Average Weekly Sales by Month and Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Month')
plt.xticks(np.arange(len(pivot_table.columns)), pivot_table.columns, rotation=45)
plt.yticks(np.arange(len(pivot_table.index)), pivot_table.index)

plt.colorbar(label='Average Weekly Sales')
plt.show()
````

##### 1. Why did you pick the specific chart?


   - This heatmap was selected to visualize the average weekly sales, providing a comprehensive view of how sales vary across both months and days of the week.





##### 2. What is/are the insight(s) found from the chart?
